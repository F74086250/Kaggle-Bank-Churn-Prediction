```python
merge=[]
for i in range(len(df_KNN)):
    sum=df_RandomForest['Exited'][i]*0.6671+df_ExtraDecisionTree['Exited'][i]*0.6149+df_SVM['Exited'][i]*0.6006+df_KNN['Exited'][i]*0.5903+df_Bagging['Exited'][i]*0.6436
    if(sum>1.55825):
        merge.append(1)
    elif(sum<=1.55825):
        merge.append(0)
    
df_RandomForest['Exited'].tolist().count(1),df_ExtraDecisionTree['Exited'].tolist().count(1),df_SVM['Exited'].tolist().count(1),df_KNN['Exited'].tolist().count(1),df_Bagging['Exited'].tolist().count(1)
```
